```python
import os
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
current_directory = os.getcwd()
ruta = current_directory + r'\DataSets-STATA'
```

```python
filename = os.path.join(ruta, 'WAGE1.DTA')
wage1 = pd.read_stata(filename, preserve_dtypes=False)
wage1.describe()
```

```python
X = wage1['educ']
y = wage1['wage']

# Add a constant (intercept) to the model
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()
print(model.summary())
print(model.resid)
```
