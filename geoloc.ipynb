```python

import pandas as pd
from geopy.geocoders import Nominatim

# Load the CSV file
input_file = r'C:\Users\franc\OneDrive - Delft University of Technology\Desktop\geoloc.csv'  # Update this with your input CSV file name
output_file = r'C:\Users\franc\OneDrive - Delft University of Technology\Desktop\geoloc2.csv'  # Name for the output CSV file

# Read the CSV file into a DataFrame
df = pd.read_csv(input_file)

# Create a new column for latitude and longitude
df['latitude'] = None
df['longitude'] = None

# Function to get coordinates using Nominatim
def get_coordinates(zip_code):
    geolocator = Nominatim(user_agent="your_app_name")
    location = geolocator.geocode(f"Spain {zip_code}")
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Iterate over the ZIP codes and get coordinates
for index, row in df.iterrows():
    zip_code = row['ZIP Code']  # Assuming the ZIP code column is named 'ZIP Code'
    lat, lon = get_coordinates(zip_code)
    df.at[index, 'latitude'] = lat
    df.at[index, 'longitude'] = lon

# Save the updated DataFrame to a new CSV file
df.to_csv(output_file, index=False)
```
